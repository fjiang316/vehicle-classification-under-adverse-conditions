```python
import os
import shutil
import random

src_path = path
dst_path = os.path.join(os.getcwd(), "normal_img/helicopter")
os.makedirs(dst_path, exist_ok=True)

# Recursively find all files
all_files = []
for root, dirs, files in os.walk(src_path):
    for f in files:
        all_files.append(os.path.join(root, f))

print(f"Total files found: {len(all_files)}")

# Select up to 200 files
num_to_select = min(300, len(all_files))
selected_files = random.sample(all_files, num_to_select)

# Move selected files
for f in selected_files:
    shutil.move(f, os.path.join(dst_path, os.path.basename(f)))

print(f"Moved {len(selected_files)} files to {dst_path}")
```

```python
def download_from_kaggle_and_select(kaggle_path, save_path, num_files):
    local_path = kagglehub.dataset_download(kaggle_path)
    dst_path = os.path.join(os.getcwd(), save_path)
    os.makedirs(dst_path, exist_ok=True)

    # Recursively find all files
    all_files = []
    for root, dirs, files in os.walk(local_path):
        for f in files:
            #print(root, dirs, f)
            if f.lower().endswith((".png", ".jpg", ".jpeg", ".bmp", ".gif", ".webp")):
                all_files.append(os.path.join(root, f))
    
    print(f"Total files found: {len(all_files)}")
    
    # Select files
    num_to_select = min(num_files, len(all_files))
    selected_files = random.sample(all_files, num_to_select)
    
    # Move selected files
    for f in selected_files:
        shutil.move(f, os.path.join(dst_path, os.path.basename(f)))
    
    print(f"Moved {len(selected_files)} files to {dst_path}")
```

```python
# Download latest version
categories_source = {"bus": "samuelayman/bus-dataset", \
                     #"car":"kshitij192/cars-image-dataset",\
                    "bicycle": "dataclusterlabs/bicycle-image-dataset-vehicle-dataset",\
                    #"boat": "clorichel/boat-types-recognition",\
                     #"truck": "mmohaiminulislam/vehicles-image-dataset",
                    "motorcycle": "zwartfreak/bike-images"}
for category, path in categories_source.items():
    download_from_kaggle_and_select(path, f"normal_img/{category}", 300)
```
