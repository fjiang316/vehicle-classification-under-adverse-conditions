```python
import os
from PIL import Image, ImageFilter

input_folder = "normal_img"
output_folder = "blurred_images_pillow"

os.makedirs(output_folder, exist_ok=True)

# adjust radius for stronger/weaker blur
BLUR_RADIUS = 8

for category in os.listdir(input_folder):
    category_path = os.path.join(input_folder, category)

    # Only process subdirectories
    if os.path.isdir(category_path) and category != "helicopter":
        # Create corresponding category folder in output
        output_category_path = os.path.join(output_folder, category)
        os.makedirs(output_category_path, exist_ok=True)

        # Loop through images inside this category folder
        for filename in os.listdir(category_path):
            if filename.lower().endswith((".png", ".jpg", ".jpeg", ".bmp", ".gif", ".webp")):
                input_path = os.path.join(category_path, filename)
                output_path = os.path.join(output_category_path, filename)

                try:
                    # Convert image to RGB to avoid "wrong mode"
                    img = Image.open(input_path).convert("RGB")

                    blurred = img.filter(ImageFilter.GaussianBlur(radius=BLUR_RADIUS))
                    blurred.save(output_path)

                    # print(f"Saved: {output_path}")

                except Exception as e:
                    print(f"Skipping {input_path} due to error: {e}")

print("Done (Pillow)!")   
```

```python
import os
import cv2

input_folder = "normal_img"
output_folder = "blurred_cv2"

os.makedirs(output_folder, exist_ok=True)

# kernel size must be odd numbers (e.g., 5, 15, 31)
KERNEL_SIZE = (21, 21)

for category in os.listdir(input_folder):
    category_path = os.path.join(input_folder, category)

    # Only process subdirectories
    if os.path.isdir(category_path):
        # Create corresponding category folder in output
        output_category_path = os.path.join(output_folder, category)
        os.makedirs(output_category_path, exist_ok=True)

        # Loop through images inside this category folder
        for filename in os.listdir(category_path):
            if filename.lower().endswith((".png", ".jpg", ".jpeg", ".bmp", ".gif", ".webp")):
                input_path = os.path.join(category_path, filename)
                output_path = os.path.join(output_category_path, filename)

                try:
                    # Convert image to RGB to avoid "wrong mode"
                    img = cv2.imread(input_path)
                    blurred = cv2.GaussianBlur(img, KERNEL_SIZE, 0)
                    cv2.imwrite(output_path, blurred)

                    # print(f"Saved: {output_path}")

                except Exception as e:
                    print(f"Skipping {input_path} due to error: {e}")

print("Done (OpenCV)!")   
```

```python
# blur by batch
input_folder = "normal_img"
output_folder = "blurred_pillow"

os.makedirs(output_folder, exist_ok=True)

# adjust radius for stronger/weaker blur
BLUR_RADIUS = 5

for filename in os.listdir(input_folder):
    if filename.lower().endswith((".png", ".jpg", ".jpeg", ".bmp", ".gif", ".webp")):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        img = Image.open(input_path)
        blurred = img.filter(ImageFilter.GaussianBlur(radius=BLUR_RADIUS))
        blurred.save(output_path)

        print(f"Saved: {output_path}")

print("Done (Pillow)!")
```
